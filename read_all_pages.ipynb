# read data from all pages

```python
import csv
import requests
from bs4 import BeautifulSoup
import pandas as pd

link_list = []
location_list = []
description_list = []
company_name_list = []
job_title_list = []

response = requests.get('https://wuzzuf.net/search/jobs/?a=hpb%7Cspbg&q=data%20science&start=0')
soup = BeautifulSoup(response.text, 'html.parser')

link = soup.find_all('a', attrs = {'class':'css-o171kl'} )
location = soup.find_all('span', attrs = {'class' : 'css-5wys0k'})
company_name = soup.find_all('a', attrs = {'class' : 'css-17s97q8'})
job_title = soup.find_all('h2' , attrs = {'class':'css-m604qf'})
    
    
for page in range(0, 40):
    
    for i in range(len(job_title)):
    
        link_list.append('https://wuzzuf.net/' + link[i].get('href'))
        location_list.append(location[i].text)
        company_name_list.append(company_name[i].text)
        job_title_list.append(job_title[i].text)

        #print('job number: ', i , '\n')
        print('job title: ' ,job_title_list[i] , '\n')
        print('Company Name: ' ,company_name_list[i] , '\n')
        print('Company Location: ' ,location_list[i] , '\n')
        print('Job Link: ' ,link_list[i] , '\n')
        print( '-'*100)
        
    res = requests.get('https://wuzzuf.net/search/jobs/?a=hpb%7Cspbg&q=data%20science&start='+str(page))
    soup = BeautifulSoup(res.text, 'html.parser')
```
