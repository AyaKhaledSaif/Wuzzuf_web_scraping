# read data from all pages and write into csv file

```python
import csv
import requests
from bs4 import BeautifulSoup
import pandas as pd

link_list = []
location_list = []
description_list = []
company_name_list = []
job_title_list = []

response = requests.get('https://wuzzuf.net/search/jobs/?a=hpb%7Cspbg&q=data%20science&start=0')
soup = BeautifulSoup(response.text, 'html.parser')

link = soup.find_all('a', attrs = {'class':'css-o171kl'} )
location = soup.find_all('span', attrs = {'class' : 'css-5wys0k'})
company_name = soup.find_all('a', attrs = {'class' : 'css-17s97q8'})
job_title = soup.find_all('h2' , attrs = {'class':'css-m604qf'})
    
    
for page in range(0, 40):
    
    for i in range(len(job_title)):
    
        link_list.append('https://wuzzuf.net/' + link[i].get('href'))
        location_list.append(location[i].text)
        company_name_list.append(company_name[i].text)
        job_title_list.append(job_title[i].text)
        
    res = requests.get('https://wuzzuf.net/search/jobs/?a=hpb%7Cspbg&q=data%20science&start='+str(page))
    soup = BeautifulSoup(res.text, 'html.parser')
        
```

```python
zip_list = zip(job_title_list, company_name_list, location_list, link_list)

with open('mine_ass.csv', 'w', newline='', encoding='UTF8') as f:
    writer = csv.DictWriter(f, fieldnames=['Job Title', 'Company Name', 'Company Location', 'Job Link'])
    writer.writeheader()
    for title, company_name, location, link in zip_list:
        
        writer.writerow({'Job Title': title , 'Company Name': company_name , 'Company Location': location  , 'Job Link': link })
```
